```python
# 1. Load your data
df_train = pd.read_csv('/home/matus/NLPD_18/part1/outputs/output_train.csv')
df_valid = pd.read_csv('/home/matus/NLPD_18/part1/outputs/output_valid.csv')

# 2. Prepare SpaCy training format
# Map label_binary into textcat "cats" dictionary
train_data = [
    (row['statement'], {"cats": {"POS": row['label_binary'] == 1, "NEG": row['label_binary'] == 0}})
    for _, row in df_train.iterrows()
]

dev_data = [
    (row['statement'], {"cats": {"POS": row['label_binary'] == 1, "NEG": row['label_binary'] == 0}})
    for _, row in df_valid.iterrows()
]

print(f"Loaded {len(train_data)} training samples and {len(dev_data)} validation samples.")
```
